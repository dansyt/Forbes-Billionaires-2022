```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patheffects as path_effects
import plotly.express as px

```

```python
df = pd.read_csv("/kaggle/input/forbes-billionaires-and-companies-2022/forbes_dataset_2022/forbes_billionaires_2022.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
TT=df.iloc[0:101,0:]          #For top 50 Companies
```

```python
fig = px.histogram(TT, x="country",template='plotly_dark',text_auto=True,color="country")
fig.update_layout(
    xaxis_title_text='Name of Country', # xaxis label
    yaxis_title_text='Count of Billionaires', # yaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)
fig.show()
```

```python
print(df.iloc[:10, :2])
```

```python
T1=df.iloc[0:10,0:]          #For top 50 Companies
```

```python
fig = px.line(T1, x="personName", y="finalWorth",template='plotly_dark')
fig.show()
```

```python
count = df["gender"].value_counts()

# Membuat pie chart beserta dengan besaran persennya
plt.pie(count, labels=count.index, autopct='%1.1f%%')

# Menampilkan grafik
plt.show()
```

```python
fig = px.histogram(TT, x="category",template='plotly_dark',text_auto=True,color="category")
fig.update_layout(
    xaxis_title_text='Name of Country', # xaxis label
    yaxis_title_text='Count of Billionaires', # yaxis label
    bargap=0.2, # gap between bars of adjacent location coordinates
    bargroupgap=0.1 # gap between bars of the same location coordinates
)
fig.show()
```

```python
px.scatter(df.sort_values(by = 'finalWorth', ascending = False)[:10], x = 'personName', y = 'finalWorth', template = 'plotly_dark', size = 'finalWorth', color = 'source', opacity = 0.85)
```

```python
fig = px.histogram(df, x = 'age', template = 'plotly_dark', color = 'age', opacity = 0.9)
fig.show()
```

```python
fig = px.bar(df.sort_values('age',ascending = True)[:10], x = 'personName', y = 'finalWorth', template = 'plotly_dark', color = 'age', opacity = 0.8)
fig.show()
```

```python
fig = px.bar(df.sort_values('age',ascending = False)[:10], x = 'personName', y = 'finalWorth', template = 'plotly_dark', color = 'age', opacity = 0.8)
fig.show()
```

```python
country_networth = df.groupby('country').sum()[['finalWorth']].sort_values('finalWorth', ascending = False).reset_index().head(10)
fig=px.scatter(country_networth, x ='country', y = 'finalWorth', template = 'plotly_dark', color = 'country', size = 'finalWorth' , opacity = 0.85)
fig.show()
```
